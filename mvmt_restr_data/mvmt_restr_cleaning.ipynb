```python
# imports
import pandas as pd
import numpy as np
import datetime
```

```python
mvmt_raw = pd.read_csv("http://api.qdatum.io/v1/pull/4?format=tsv", sep='\t')
mvmt_raw.head()
```

```python
mvmt_interpolate = pd.DataFrame()
# returns list of dates given a start/end date and country/region of restriction
def daterange(country, region, start, end):
    if end == "n/a":
        end = datetime.date.today()
    numdays = (end-start).days
    return [(country, region, start + datetime.timedelta(days=x), 1) for x in range(0, numdays)]

for r in mvmt_raw.iterrows():
    # skip potential mvmt restrictions
    if r[1][6] == "Potential":
        continue
    start = datetime.datetime.strptime(r[1][13], '%Y-%m-%d')
    end = r[1][14]
    country, region = r[1][2], r[1][3]
    if end == "n/a":
        end = datetime.datetime.now()
    else:
        end = datetime.datetime.strptime(end, '%Y-%m-%d')
    tmp = pd.DataFrame(daterange(country,region,start,end))
    mvmt_interpolate = mvmt_interpolate.append(tmp)
```

```python
mvmt_interpolate.columns = ["Country", "Region", "Date", "IfRestrict"]
mvmt_interpolate.head()
```
